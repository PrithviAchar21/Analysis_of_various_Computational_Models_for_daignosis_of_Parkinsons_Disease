```python
import numpy as np
import pandas as pd
iris = pd.read_csv('Parkinsondisease.csv')
iris.head()
```

```python
x = iris.drop(['status','name'],axis=1)
y = iris['status']
```

```python
pd.DataFrame(x,columns=iris.columns[:-1])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

# svm


```python
from sklearn import svm
clf = svm.SVC(kernel='linear',C=0.01)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=10)
rf_clf.fit(x_train,y_train)
y_rf=rf_clf.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
print(confusion_matrix(y_test,y_rf))
print(accuracy_score(y_test,y_rf))
print(classification_report(y_test,y_rf))
```

# GradientBoost

```python
from sklearn.ensemble import GradientBoostingClassifier
grad_boost = GradientBoostingClassifier(learning_rate=0.1)
grad_boost.fit(x_train,y_train)
y_grad = grad_boost.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
print(confusion_matrix(y_test,y_grad))
print(accuracy_score(y_test,y_grad))
print(classification_report(y_test,y_grad))
```

# Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
ada_boost = AdaBoostClassifier()
ada_boost.fit(x_train,y_train)
y_ada = ada_boost.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
print(confusion_matrix(y_test,y_ada))
print(accuracy_score(y_test,y_ada))
print(classification_report(y_test,y_ada))
```
